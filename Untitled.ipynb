```python
def nulos_por_grupo(df, grupo_cols, target_col):
    agrupar_nulos = (
        df.groupby(grupo_cols, observed=True)[target_col]
        .apply(lambda x: x.isnull().mean() * 100)
        .reset_index(name=f'porcentaje_nulos_{target_col}')
    )
    return agrupar_nulos
```

```python
print(df["resource_allocation"].unique())
nulos_por_grupo(df, ["company_type", "wfh_setup_available"],"resource_allocation")
```

```python
nulos_por_grupo(df, ["manager_support_score"],"resource_allocation")
```

# Distribucion de los datos y outliers

Ahora que no hay datos faltantes, miramos un poco las distribuciones de las distintas variables

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

print("Numerical columns:", numerical_cols)
print("Categorical columns:", categorical_cols)
```

Primer vistazo general para visualizar rapido las columnas

```python
for col in numerical_cols:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    global_min = min(df[col].min(), df_na[col].min())
    global_max = max(df[col].max(), df_na[col].max())
    
    # Define bins (adjust number of bins as needed)
    bins = np.linspace(global_min, global_max, 30) 
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=False, bins=bins, ax=axes[0])
    sns.histplot(df_na[col], kde=False, bins=bins, ax=axes[1])
    plt.show()
```

Ya de esta observación se puede ver que los atributos del dataframe no tienen outliers. 
Las variables estan acotadas entre valores "razonables".
Algunos parecen ser puntuaciones entre 0 y 5 (10) como resource_allocation y designation


```python
values = df.mental_fatigue_score.unique()
values.sort()
print("Valores de mental_fatigue: ")
print(values)
values = df.burn_rate.unique()
values.sort()
print("Valores de burn_rate: ")
print(values)
```

Las columnas burn_rate y mental_fatigue_score son en realidad variables discretas que pueden tomar valores entre 0 y 1 (resp. 0 y 10) y valores cada 0.01 (resp. 0.1)

```python
plt.figure(figsize=(6, 4))
sns.histplot(df["burn_rate"], kde=True, bins=20)
```

- work_life_balance_score

Todas toman valores discretos y contiguos, se descarta la existencai de outliers

Tampoco se observan outliers en las horas semanales trabajadas que estan acotadas entre 35 y 60.
En este caso parecen haber dos distribuciones uniformes (35/45 y 45/60)
Miramos en detalle la columna

```python
plt.figure(figsize=(6, 4))
sns.histplot(df[df.work_hours_per_week < 47]["work_hours_per_week"], kde=False, bins=10)
sns.histplot(df[df.work_hours_per_week >= 47]["work_hours_per_week"], kde=False, bins=12)
mean_lt_47 = df[df.work_hours_per_week < 47]["work_hours_per_week"].mean()
mean_ge_47 = df[df.work_hours_per_week >= 47]["work_hours_per_week"].mean()
print(f"\nWork Hours mean (< 47hs): {mean_lt_47:.2f}")
print(f"Work Hours mean (>= 47hs): {mean_ge_47:.2f}")
```

Queremos ver que otra(s) variables son las que separan estas dos distribuciones. 

Probamos primero mirando media y mediana de las variables binarias.
Sin ser un test formal, se esperaria que si alguna de estas es la causante de las segmentación, se observen distintas medias para ambos valores

```python
for col in ["gender", "company_type", "wfh_setup_available"]:
    print(f"\nWork Hours by {col}:")
    print(df.groupby(col)["work_hours_per_week"].agg(["mean", "median", "count"]))
```

Ninguna de las tres variables parece separar las medias lo suficiente y se hallan cerca del rango medio de la distribucion completa. 
Por lo que no se les podría endilgar la separación observada.

Para explicar esta separación debemos mirar otras variables. Una posibilidad es que la cantidad de horas trabajadas dependa del nivel (desiganation)

```python
plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x="designation", y="work_hours_per_week", showmeans=True)
plt.axhline(mean_lt_47, color='red', linestyle='--', alpha=0.5, label=f'{mean_lt_47:.2f}')
plt.axhline(mean_ge_47, color='green', linestyle='--', alpha=0.5, label=f'{mean_ge_47:.2f}')
plt.title("Horas trabajadas x nivel de trabajo")
plt.legend()
plt.show()

plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x="manager_support_score", y="work_hours_per_week", showmeans=True)
plt.axhline(mean_lt_47, color='red', linestyle='--', alpha=0.5, label=f'{mean_lt_47:.2f}')
plt.axhline(mean_ge_47, color='green', linestyle='--', alpha=0.5, label=f'{mean_ge_47:.2f}')
plt.title("Horas trabajadas x nivel de trabajo")
plt.legend()
plt.show()
```
